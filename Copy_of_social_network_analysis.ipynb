```
import networkx as nx
import matplotlib.pyplot as plt
```

```
G_symmetric = nx.Graph()


#ID0
G_symmetric.add_edge('James','David',weight=1)
G_symmetric.add_edge('James','George',weight=1)



#ID1
G_symmetric.add_edge('David','George', weight=1)


#ID2
G_symmetric.add_edge('George','Ronald', weight=2)
G_symmetric.add_edge('George','John', weight=2)
G_symmetric.add_edge('George','Richard', weight=3)



#ID3
G_symmetric.add_edge('Ronald','Daniel', weight=4)
G_symmetric.add_edge('Ronald','Kenneth', weight=5)
G_symmetric.add_edge('Ronald','Anthony', weight=5)



#ID4
G_symmetric.add_edge('John','Robert', weight=6)
G_symmetric.add_edge('John','Charles', weight=4)
G_symmetric.add_edge('John','Paul', weight=5)
G_symmetric.add_edge('John','Mark', weight=7)
G_symmetric.add_edge('John','Kevin', weight=4)
G_symmetric.add_edge('John','Edward', weight=5)




#ID5
G_symmetric.add_edge('Richard','Joseph', weight=5)
G_symmetric.add_edge('Richard','Michael', weight=3)
G_symmetric.add_edge('Richard','Jason', weight=5)




#ID6
G_symmetric.add_edge('Daniel','Kenneth', weight=3)
G_symmetric.add_edge('Daniel','Anthony', weight=4)



#ID7
G_symmetric.add_edge('Kenneth','Anthony', weight=4)



#ID9
G_symmetric.add_edge('Robert','Charles', weight=2)
G_symmetric.add_edge('Robert','Paul', weight=4)
G_symmetric.add_edge('Robert','Mark', weight=7)
G_symmetric.add_edge('Robert','Kevin', weight=2)
G_symmetric.add_edge('Robert','Edward', weight=4)




#ID10
G_symmetric.add_edge('Charles','Paul', weight=5)
G_symmetric.add_edge('Charles','Mark', weight=5)
G_symmetric.add_edge('Charles','Kevin', weight=4)
G_symmetric.add_edge('Charles','Edward', weight=3)




#ID11
G_symmetric.add_edge('Paul','Mark', weight=3)
G_symmetric.add_edge('Paul','Kevin', weight=4)



#ID12
G_symmetric.add_edge('Mark','Edward',weight=3)



#ID15
G_symmetric.add_edge('Joseph','Michael', weight=3)
G_symmetric.add_edge('Joseph','Jason', weight=4)



#ID16
G_symmetric.add_edge('Michael','Jason', weight=5)



#ID17
G_symmetric.add_edge('Jason','Kenneth', weight=1)



#ID16
G_symmetric.add_edge('Michael','Daniel', weight=6)
G_symmetric.add_edge('Michael','Kenneth', weight=7)
G_symmetric.add_edge('Michael','Anthony', weight=8)
G_symmetric.add_edge('Michael','Robert', weight=9)
G_symmetric.add_edge('Michael','Paul', weight=11)
G_symmetric.add_edge('Michael','Kevin', weight=13)
G_symmetric.add_edge('Michael','Edward', weight=14)
G_symmetric.add_edge('Michael','Charles', weight=10)


nx.clustering(G_symmetric)

pos = nx.spring_layout(G_symmetric)
betCent = nx.betweenness_centrality(G_symmetric, normalized=True, endpoints=True)

node_size =  [v * 5000 for v in betCent.values()]
plt.figure(figsize=(10,10))
nx.draw_networkx(G_symmetric, pos=pos,
                  nodelist=['James','David','George'],
                       node_color='red',
                  node_size=node_size )

nx.draw_networkx(G_symmetric, pos=pos,
                  nodelist=['Ronald','John','Richard'],
                       node_color='blue',
                  node_size=node_size )

nx.draw_networkx(G_symmetric, pos=pos,
                  nodelist=['Daniel','Kenneth','Anthony'],
                       node_color='yellow',
                  node_size=node_size )


nx.draw_networkx(G_symmetric, pos=pos,
                  nodelist=['Robert','Charles','Paul'],
                       node_color='Orange',
                  node_size=node_size )


nx.draw_networkx(G_symmetric, pos=pos,
                  nodelist=['Mark','Kevin','Edward'],
                       node_color='Green',
                  node_size=node_size )


nx.draw_networkx(G_symmetric, pos=pos,
                  nodelist=['Joseph','Michael','Jason'],
                       node_color='cyan',
                  node_size=node_size )

```
